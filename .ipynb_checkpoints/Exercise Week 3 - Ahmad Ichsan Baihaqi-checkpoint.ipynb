```python
### Exercise Week 3

### Author: Ahmad Ichsan Baihaqi
### Email: ahmadichsanbaihaqi@gmail.com

def question_one():
    print("QUESTION ONE")
    
    def substring_count(input_string, key):
        counter = 0

        key_length = len(key)

        for index, item in enumerate(input_string):
            start = index
            end = start + key_length
            content = input_string[start : end]

            if (content == key):
                counter += 1
        return counter

    def printing(_key, _input, _result):
        print(_key + " occurance in " + _input + ' is ' + str(_result))
        
    # Output: 2
    input1 = 'ABCDCDC'
    key1 = 'CD'
    result1 = substring_count(input1, key1)
    printing(key1, input1, result1)

    # Output: 2
    input2 = 'ABCDCDC'
    key2 = 'CDC'
    result2 = substring_count(input2, key2)
    printing(key2, input2, result2)

    # Output: 3
    input3 = 'DEDFDEDEDEG'
    key3 = 'DED'
    result3 = substring_count(input3, key3)
    printing(key3, input3, result3)
    
question_one()


print('==============================================')


def question_two():
    print("QUESTION TWO")
        
    # use recursive
    def diff_letters(input, _diff):
            
        for item in input:
            splitted = input.split(item)
            filterred = [i for i in splitted if len(i) != 0]
            
            _diff += 1
            
            if (len(filterred) != 0):
                _diff = diff_letters(''.join(filterred), _diff)
            
            break
        
        return _diff
    
    def printing(_input, _result):
        print("different letters in " + _input + ' is ' + str(_result))
        
    # Output: 2
    input1 = 'AAAAAAAAAB'
    result1 = diff_letters(input1, 0)
    printing(input1, result1)
    
    # Output: 5
    input2 = 'ABCDAAAABBCCCE'
    result2 = diff_letters(input2, 0)
    printing(input2, result2)

question_two()


print('==============================================')


def question_three():
    print("QUESTION THREE")
    
    import re

    def distance(input):
        # The pattern -?\d+ means match a group of digits (\d+) that can possibly be preceded by a minus sign (-?).
        numbers = re.compile('-?\d+')
        extracted_list = list(map(int, numbers.findall(input)))
        
        x1 = extracted_list[0]
        y1 = extracted_list[1]
        x2 = extracted_list[2]
        y2 = extracted_list[3]
        
        first_point = x1 - x2
        first_point_square = first_point ** 2
        
        second_point = y1 - y2
        second_point_square = second_point ** 2
        
        result = (first_point_square + second_point_square) ** 0.5
        return result
    
    # Output: 4.242640687119285
    input1 = '(0,-1), (-3,2)'
    result1 = distance(input1)
    print('The distance of ' + input1 + ' is ' + str(result1))

question_three()
```
